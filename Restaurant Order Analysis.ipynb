```python
USE restaurant_db;
```

```python
-- View menu_items table
SELECT * FROM menu_items;
```

```python
-- View order_Details table
SELECT * FROM order_details;
```

```python
-- Total Number of items on the menu
SELECT COUNT(menu_item_id) AS total_items
FROM menu_items;
```

```python
-- leat and most expensive items on menu
SELECT MIN(price) AS least_expensive, MAX(price) AS most_expensive
FROM menu_items;
```

```python
-- Total Italian Dishes on Mneu
SELECT COUNT(*) AS total_italian_dishes
FROM menu_items
WHERE category = 'Italian'
```

```python
-- Least and Most Expensive Italian Dishes
SELECT MIN(price) AS least_expensive_italian,
MAX(price) AS most_expensive_italian
FROM menu_items
WHERE category = 'Italian'
```

```python
-- Total Dishes in Each Category
SELECT category, COUNT(*) AS total_dishes
FROM menu_items
GROUP BY category;
```

```python
-- Average Dish Price in each category
SELECT category, AVG(price) AS avg_price
FROM menu_items
GROUP BY category;
```
