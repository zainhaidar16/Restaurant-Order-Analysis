```python
import mysql.connector
import pandas as pd

# Database conn details
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'root',
    'database': 'restaurant_db',
}

def execute_sql_query(sql_query, conn):
    try:
        # Execute the SQL query and return the result as a DataFrame
        return pd.read_sql_query(sql_query, conn)

    except mysql.connector.Error as err:
        print(f"Error: {err}")
        return None

# Establish a conn to the database
conn = mysql.connector.connect(**db_config)

if conn.is_connected():
    print("Connected to the database")
```

# Explore the items table
#### The objective is to better understand the items table by finding the number of rows in the table, the least and most expensive items, and the item prices within each category.

```python
# view items from menu_items
sql_query = 'SELECT * FROM menu_items'
df = execute_sql_query(sql_query,conn)
df
```

```python
# Total Number of items on the menu
sql_query = "SELECT COUNT(menu_item_id) AS total_items FROM menu_items"
df = execute_sql_query(sql_query, conn)
df
```

```python
# least and most expensive items on menu
sql_query = "SELECT MIN(price) AS least_expensive, MAX(price) AS most_expensive FROM menu_items"
df = execute_sql_query(sql_query,conn)
df
```

```python
# Total Italian Dishes on Mneu
sql_query = "SELECT COUNT(*) AS total_italian_dishes FROM menu_items WHERE category = 'Italian'"
df = execute_sql_query(sql_query,conn)
df
```

```python
# Least and Most Expensive Italian Dishes
sql_query = "SELECT MIN(price) AS least_expensive_italian, MAX(price) AS most_expensive_italian FROM menu_items WHERE category = 'Italian'"
df = execute_sql_query(sql_query,conn)
df
```

```python
# Total Dishes in Each Category
sql_query = "SELECT category, COUNT(*) AS total_dishes FROM menu_items GROUP BY category"
df = execute_sql_query(sql_query,conn)
df
```

```python
# Average Dish Price in each category
sql_query = "SELECT category, AVG(price) AS avg_price FROM menu_items GROUP BY category"
df = execute_sql_query(sql_query,conn)
df
```

# Explore the orders table
#### The objective is to better understand the orders table by finding the date range, the number of items within each order, and the orders with the highest number of items.

```python
# view order_details table
sql_query = "SELECT * FROM order_details LIMIT 10"
df = execute_sql_query(sql_query, conn)
df
```

```python
# date rnage of order_details table
sql_query = "SELECT MIN(order_date) as first_order_date, MAX(order_date) AS last_order_date FROM order_details"
df = execute_sql_query(sql_query, conn)
df
```

```python
# How many orders were made within this date range?
sql_query = """
SELECT COUNT(*) AS total_orders
FROM order_details
WHERE order_date >= (SELECT MIN(order_date) FROM order_details)
  AND order_date <= (SELECT MAX(order_date) FROM order_details);
"""
df = execute_sql_query(sql_query, conn)
df
```

```python
# Which orders had the most number of items?
sql_query = """
SELECT order_id, COUNT(*) AS item_count
FROM order_details
GROUP BY order_id
ORDER BY item_count DESC
LIMIT 1;
"""
df = execute_sql_query(sql_query,conn)
df
```

```python
# How many orders had more than 12 items?
sql_query ="""
SELECT order_id, COUNT(*) AS total_items
FROM order_details
GROUP BY order_id
HAVING COUNT(*) > 12; 
"""
df = execute_sql_query(sql_query,conn)
df
```

# Analyze customer behavior
#### The objective is to combine the items and orders tables, find the least and most ordered categories, and dive into the details of the highest spend orders.

```python
# Combine the menu_items and order_details tables into a single table
sql_query = """
CREATE TABLE combined_table AS
SELECT 
    order_details.order_id,
    order_details.order_date,
    order_details.order_time,
    order_details.item_id AS order_item_id,
    menu_items.item_name,
    menu_items.category,
    menu_items.price
FROM order_details
JOIN menu_items ON order_details.item_id = menu_items.menu_item_id;
"""
df = execute_sql_query(sql_query,conn)
```

```python
# What were the least and most ordered items? What categories were they in?
sql_query = """
WITH ItemCounts AS (
    SELECT 
        ct.order_item_id AS item_id,
        ct.item_name,
        ct.category,
        COUNT(*) AS order_count
    FROM combined_table ct
    GROUP BY ct.order_item_id, ct.item_name, ct.category
)
SELECT 
    item_id,
    item_name,
    category,
    order_count
FROM ItemCounts
WHERE order_count = (SELECT MIN(order_count) FROM ItemCounts)
   OR order_count = (SELECT MAX(order_count) FROM ItemCounts);
"""

df = execute_sql_query(sql_query,conn)
df
```

```python
# What were the top 5 orders that spent the most money?
sql_query = """ 
WITH OrderTotalCost AS (
    SELECT
        ct.order_id,
        SUM(mi.price) AS total_cost
    FROM combined_table ct
    JOIN menu_items mi ON ct.order_item_id = mi.menu_item_id
    GROUP BY ct.order_id
)
SELECT
    otc.order_id,
    otc.total_cost
FROM OrderTotalCost otc
ORDER BY otc.total_cost DESC
LIMIT 5;
"""
df = execute_sql_query(sql_query, conn)
df
```

```python
# View the details of the highest spend order. Which specific items were purchased?
sql_query = """
WITH OrderTotalCost AS (
    SELECT
        ct.order_id,
        SUM(mi.price) AS total_cost
    FROM combined_table ct
    JOIN menu_items mi ON ct.order_item_id = mi.menu_item_id
    GROUP BY ct.order_id
)
SELECT
    ct.order_id,
    ot.total_cost,
    ct.order_item_id,
    ct.item_name,
    ct.category,
    ct.price AS item_price
FROM OrderTotalCost ot
JOIN combined_table ct ON ot.order_id = ct.order_id
WHERE ot.total_cost = (SELECT MAX(total_cost) FROM OrderTotalCost);
"""
df = execute_sql_query(sql_query, conn)
df
```

```python
# View the details of the top 5 highest spend orders
sql_query = """
WITH OrderTotalCost AS (
    SELECT
        ct.order_id,
        SUM(mi.price) AS total_cost
    FROM combined_table ct
    JOIN menu_items mi ON ct.order_item_id = mi.menu_item_id
    GROUP BY ct.order_id
)
SELECT
    ot.order_id,
    ot.total_cost,
    ct.order_item_id,
    ct.item_name,
    ct.category,
    ct.price AS item_price
FROM OrderTotalCost ot
JOIN combined_table ct ON ot.order_id = ct.order_id
ORDER BY ot.total_cost DESC
LIMIT 5;
"""
df = execute_sql_query(sql_query,conn)
df
```
